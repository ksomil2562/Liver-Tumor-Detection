```python
from glob import glob
import shutil
import os
import dicom2nifti
import nibabel as nib
import numpy as np
```

```python
patient_list = glob(in_path+'/*')
patient_list.sort()
print(patient_list[0])
```

```python
for patient in glob(in_path + '/*'):
    patient_name = os.path.basename(os.path.normpath(patient))
    number_folder = int(len(glob(patient+'/*'))/64)

    for i in range(number_folder):
        output_path_name = os.path.join(out_path,patient_name + '_' + str(i))
        os.mkdir(output_path_name)
        #print(output_path_name)
        for i ,file in enumerate(glob(patient+'/*')):
            if i == 64 + 1:
                break
            shutil.move(file,output_path_name)
```

```python
in_path_images = '/Users/pratyush/Downloads/liversegmentation/output/images/*'
in_path_labels = '/Users/pratyush/Downloads/liversegmentation/output/labels/*'
out_path_images = '/Users/pratyush/Downloads/liversegmentation/nifti_files/images'
out_path_labels = '/Users/pratyush/Downloads/liversegmentation/nifti_files/labels'

list_images = glob(in_path_images)
list_images.sort()
list_labels = glob(in_path_labels)
list_labels.sort()
list_labels
```

```python
import dicom2nifti.settings as settings

# Disable validation
settings.validate_slice_increment = False
settings.validate_orthogonal = False

for patient in list_labels:
    patient_name = os.path.basename(os.path.normpath(patient))
    dicom2nifti.dicom_series_to_nifti(patient,os.path.join(out_path_labels,patient_name+'.nii'))
```
